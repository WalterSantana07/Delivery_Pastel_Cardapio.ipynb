```python
# Boas-vindas ao cliente
print("Bem-vindes à Pastéis Delícia!")
nome_cliente = input("Digite o seu nome: ").capitalize()
print("Bom ter você com a gente,", nome_cliente, "!")

#Variavel e Dicionário!!
pasteis = {
    "00": {"nome": "Pastel de carne", "valor": 12},
    "01": {"nome": "Pastel de queijo", "valor": 14},
    "02": {"nome": "Pastel de catupiry", "valor": 17},
    "03": {"nome": "Pastel de siri", "valor": 32},
    "04": {"nome": "Pastel de camarao", "valor": 35},
    "05": {"nome": "Pastel de jacaré", "valor": 39},
    "06": {"nome": "Pastel de bode", "valor": 25},
    "07": {"nome": "Pastel de carneiro", "valor": 25},
    "08": {"nome": "Pastel de traíra", "valor": 45},
    "09": {"nome": "Pastel de piaba", "valor": 35}
}

#Estrura de Funções, Formatações, Condicões, Repetições!!!
def formatar_valor(valor):
    return f"R${valor:,.2f}".replace(",", ".")

def verificar_cardapio():
    print("Cardápio Disponível:")
    for codigo, pastel_info in pasteis.items():
        valor = pastel_info['valor']
        print(f"Código: {codigo} - {pastel_info['nome']} - {formatar_valor(valor)}/qtd")

def fazer_pedido():
    pedido = {}
    while True:
        codigo = input("Digite o código do pastel desejado (ou '0' para finalizar o pedido): ")
        if codigo == '0':
            break

        if codigo in pasteis:
            pastel_info = pasteis[codigo]
            qtd = int(input("Digite quantos pasteis deseja: "))
            if qtd > 0:
                if codigo in pedido:
                    pedido[codigo] += qtd
                else:
                    pedido[codigo] = qtd
            else:
                print("Quantidade inválida. Digite um número maior que zero.")
        else:
            print("Código de pastel inválido.")

#Usando a Condição If e a Função Sum para fazer as operações Matemáticas!!!
    if pedido:
        valor_total = sum(pasteis[codigo]['valor'] * qtd for codigo, qtd in pedido.items())
        taxa_entrega = 10.00
        valor_total += taxa_entrega

#Confirmando o Pedido do Cliente!!!
        print("\nVocê pediu:")
        for codigo, qtd in pedido.items():
            pastel_info = pasteis[codigo]
            print(f"- {qtd} {pastel_info['nome']}")

        print(f"\nValor dos pasteis: {formatar_valor(valor_total)}")
        print(f"Taxa de entrega: {formatar_valor(taxa_entrega)}")
        print(f"Valor total: {formatar_valor(valor_total)}")

 #Usando o Método 'lower'para  converter os caracteres
        forma_pagamento = input("\nSelecione a forma de pagamento (pix ou cartao): ").capitalize()
        if forma_pagamento.lower() == "pix":
            print(f"\nRealize o pagamento via Pix em nossa chave celular 7777777, em nome de Pasteis Delícia, tte encaminhe o comprovante")
            print("Pagamento Realizado com sucesso!")

        elif forma_pagamento.lower() == "cartao":
            print(f"\nInsira os dados do cartão para confirmar o pedido.")
            print(f"Pagamento Realizado com sucesso!")
        else:
            print("Forma de pagamento inválida.")

 #Criando Variaveis para guardar as informações de entrega!!!
        nome_rua = input("\nInforme o nome da rua: ").capitalize()
        bairro = input("Informe o bairro: ").capitalize()
        referencia = input("Informe um ponto de referência para entrega: ").capitalize()

        print("\nA entrega será realizada em até 30 minutos na Rua", nome_rua, "no bairro", bairro, "Ponto de referencia", referencia)

        for codigo, qtd in pedido.items():
            del pasteis[codigo]
    else:
        print("\nPedido vazio. Nenhum pastel foi adicionado.")

#Criando o 'Menu' com Repetiçao 'while', 'break' e o Método 'items'!!!
while True:
    print("\nMenu:")
    print("1 - Verificar Cardápio")
    print("2 - Fazer Pedido")
    print("3 - Sair")
    opcao = input("Digite a opção desejada: ")

    if opcao == '1':
        verificar_cardapio()
    elif opcao == '2':
        fazer_pedido()
    elif opcao == '3':
        break
    else:
        print("Opção inválida.")
```
